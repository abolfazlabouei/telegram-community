### Remove "WAITING_FOR_CHECK" and extra double quotes

```python
# Specify the input and output file paths
input_file_path = "subscriptions2.txt"
output_file_path = "output.txt"

# Open the input and output files
with open(input_file_path, 'r') as input_file, open(output_file_path, 'w') as output_file:
    for line in input_file:
        # Remove "WAITING_FOR_CHECK" and extra double quotes
        modified_line = line.replace('"WAITING_FOR_CHECK"', '').replace('""', '"')
        
        # Write the modified line to the output file
        output_file.write(modified_line)

# Close the files
input_file.close()
output_file.close()
```

### Select the first 3 parts

```python

# Read the content of the output.txt file
with open("output.txt", "r") as file:
    lines = file.readlines()

# Process the lines and select the first 3 parts
data = []
for line in lines:
    parts = line.strip().split(",")
    selected_parts = parts[:3]
    data.append(selected_parts)

# Create a new DataFrame
df = pd.DataFrame(data, columns=["group_id", "number_of_member", "peer_name"])

# Optional: You can save the DataFrame to a new CSV file if needed
df.to_csv("output.csv", index=False)

# Display the new DataFrame
print(df)
```
