```python
import networkx as nx
import pandas as pd
```

```python
df = pd.read_csv("merged_df.csv")
```

```python
df
```

```python
df = df.drop(columns="post")
```

```python
df = df.iloc[:100000]
```

```python
df
```

```python
# Create an empty graph
G = nx.Graph()

# Iterate through the DataFrame to add nodes for each group
for group_id, group_data in df.groupby('group_id'):
    users = set(group_data['user_id'])
    group_name = group_data['peer_name'].iloc[0]
    G.add_node(group_id, group_name=group_name)

# Compare each group's users with the users of other groups and add edges if they have common users
group_ids = df['group_id'].unique()
for i in range(len(group_ids)):
    for j in range(i + 1, len(group_ids)):
        group_id1 = group_ids[i]
        group_id2 = group_ids[j]
        
        users1 = set(df[df['group_id'] == group_id1]['user_id'])
        users2 = set(df[df['group_id'] == group_id2]['user_id'])
        
        common_users = len(users1.intersection(users2))
        
        if common_users > 10:
            G.add_edge(group_id1, group_id2, common_users=common_users)

# Create a DataFrame to store the edges and group names
edge_data = []
for u, v, data in G.edges(data=True):
    group_name_u = G.nodes[u]['group_name']
    group_name_v = G.nodes[v]['group_name']
    common_users = data['common_users']
    edge_data.append({'group_id_1': u, 'group_name_1': group_name_u, 'group_id_2': v, 'group_name_2': group_name_v, 'common_users': common_users})

df_edges = pd.DataFrame(edge_data)

# Save the DataFrame to a CSV file
df_edges.to_csv('edge_data_with_names.csv', index=False)

```
